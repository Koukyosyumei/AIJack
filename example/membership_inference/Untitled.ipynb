```python
import numpy as np
import random
import matplotlib.pyplot as plt

import torch
import torchvision
import torchvision.transforms as transforms
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data.dataset import Dataset
from torch.utils.data import DataLoader

import copy
```

```python
# ToTensor：画像のグレースケール化（RGBの0~255を0~1の範囲に正規化）、Normalize：Z値化（RGBの平均と標準偏差を0.5で決め打ちして正規化）
transform = transforms.Compose([transforms.ToTensor()])
```

```python
trainset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
testset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)
```

```python
trainloader = torch.utils.data.DataLoader(trainset, batch_size=4, shuffle=False)
valloader = torch.utils.data.DataLoader(testset, batch_size=4, shuffle=False)
```

```python
for (xs, ys) in trainloader:
    break

for i, x in enumerate(xs):
    plt.subplot(1,4,i+1)
    plt.imshow(x.detach().numpy().T)
```

```python
transform_augmentation = transforms.Compose([
                                             transforms.RandomCrop(32, padding=4),
                                             transforms.RandomHorizontalFlip(),
                                             transforms.ToTensor(),])
```

```python
augmented_loader = copy.deepcopy(trainloader)
augmented_loader.dataset.transform = transform_augmentation

torch.manual_seed(17)

for (xs, ys) in augmented_loader:
    break

for i, x in enumerate(xs):
    plt.subplot(1,4,i+1)
    plt.imshow(x.detach().numpy().T)
```
